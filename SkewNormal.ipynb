```python
import numpy as np
from scipy.stats import skewnorm
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1)
a = 4

mean, var, skew, kurt = skewnorm.stats(a, moments='mvsk')

x = np.linspace(skewnorm.ppf(0.01, a),
                skewnorm.ppf(0.99, a), 100)
ax.plot(x, skewnorm.pdf(x, a),
       'r-', lw=5, alpha=0.6, label='skewnorm pdf')
```

```python
fig, ax = plt.subplots(1, 1)
rv = skewnorm(a)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
fig.show()
```

```python
fig, ax = plt.subplots(1, 1)

vals = skewnorm.ppf([0.001, 0.5, 0.999], a)
np.allclose([0.001, 0.5, 0.999], skewnorm.cdf(vals, a))
r = skewnorm.rvs(a, size=1000)

ax.hist(r, density=True, bins='auto', histtype='stepfilled', alpha=0.2)
ax.set_xlim([x[0], x[-1]])
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
from scipy import stats

# choose some parameters
a, loc, scale = 1.3, -0.1, 2.2
# draw a sample
sample = stats.skewnorm(a, loc, scale).rvs(1000)
```

```python
x = np.sort(sample)
```

```python
import seaborn as sns
sns.histplot(sample, bins=30, kde=True,stat='density')
ae, loce, scalee = stats.skewnorm.fit(sample)

plt.plot(x, skewnorm.pdf(x, ae, loc=loce, scale=scalee),
       'r-', lw=5, alpha=0.6, label='skewnorm pdf')

percentile = 2
percentile_value = np.percentile(sample,percentile)

plt.axvline(percentile_value,color='orange')

plt.show()
```
