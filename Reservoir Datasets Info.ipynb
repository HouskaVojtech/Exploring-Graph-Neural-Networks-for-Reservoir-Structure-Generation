```python
sns.histplot(res30['mean_rmse'], bins=100, kde=True)
plt.xlim(0.04,0.25)
plt.xlabel('Mean RMSE')
plt.savefig('images/Reservoir_Dataset/hist_res30.png')
plt.savefig('images/Reservoir_Dataset/hist_res30.pdf')
plt.savefig('images/Reservoir_Dataset/hist_res30.svg')
```

```python
import pandas as pd
df16 = pd.DataFrame({'Dataset' : ['res16'] * len(res16['mean_rmse']), 'Score' : res16['mean_rmse']})
df23 = pd.DataFrame({'Dataset' : ['res23'] * len(res23['mean_rmse']), 'Score' : res23['mean_rmse']})
df30 = pd.DataFrame({'Dataset' : ['res30'] * len(res30['mean_rmse']), 'Score' : res30['mean_rmse']})
frames = [df16, df23, df30]
df = pd.concat(frames)
sns.boxplot(x="Dataset", y="Score",
            palette=["m", "g" , "r"],
            data=df)
plt.savefig('images/Reservoir_Dataset/candle_plot_reservoirs.png')
plt.savefig('images/Reservoir_Dataset/candle_plot_reservoirs.pdf')
plt.savefig('images/Reservoir_Dataset/candle_plot_reservoirs.svg')
```
